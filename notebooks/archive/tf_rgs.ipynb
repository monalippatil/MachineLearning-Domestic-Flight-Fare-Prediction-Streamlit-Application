```python
def df_to_dataset(features, target, shuffle=True, batch_size=32):
    
  ds = tf.data.Dataset.from_tensor_slices((features.to_dict(orient='list'), target))
   # ds = tf.data.Dataset.from_tensor_slices((dict(features), target))
  if shuffle:
    ds = ds.shuffle(buffer_size=len(features))
  ds = ds.batch(batch_size)
  return ds
```

```python
train_ds = df_to_dataset(X_train, y_train)
val_ds = df_to_dataset(X_val, y_val, shuffle=False)
test_ds = df_to_dataset(X_test, y_test, shuffle=False)
```

```python
from tensorflow import feature_column
```

```python
train_ds
```

```python
df.columns
```

```python
df.info()
```

```python
feature_columns = []
```

```python
for col_name in ['isBasicEconomy','isRefundable', 'isNonStop', 'totalTravelDistance', 'flightDate_month', 'flightDate_day', 'flightDate_dow','flightDuration','flightDepartureHour','flightDepartureMinute']:
  feature_columns.append(feature_column.numeric_column(col_name))
```

```python
for col_name in ['startingAirport', 'destinationAirport','cabinCode']:
  categorical_column = feature_column.categorical_column_with_vocabulary_list(col_name, features[col_name].unique())
  indicator_column = feature_column.indicator_column(categorical_column)
  feature_columns.append(indicator_column)
```

```python
tf.random.set_seed(42)
```

```python
feature_columns[:]
```

```python
from tensorflow.keras.layers import Dense, Dropout, DenseFeatures
feature_layer = DenseFeatures(feature_columns)
```

```python
model = tf.keras.Sequential()
```

```python
layer1 = Dense(128, activation='relu')
#layer2 = Dense(128, activation='relu')
dropout = Dropout(0.2)
top_layer = Dense(1)
```

```python
model.add(feature_layer)
model.add(layer1)
#model.add(layer2)
model.add(dropout)
model.add(top_layer)
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=1e-3)
```

```python
model.compile(optimizer=optimizer, loss='mean_absolute_error', metrics=['mean_absolute_error', 'mean_squared_error'])
```

```python
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    verbose=1,
    patience=5,
    restore_best_weights=True
    )
```

```python
history = model.fit(
    train_ds,
    validation_data=val_ds,
    callbacks=[early_stopping],
    epochs=13
    )
```

```python
model.evaluate(test_ds)
```
